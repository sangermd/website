```python
# Exercise: Can you select Columns [Date, Time, T] with PT08.S2(NMHC) < 740 or NOx(GT) not in [92.0, 95.0, 153.0, 158.0]?
# condition 1 : PT08.S2(NMHC) < 740
# condition 2: NOx(GT) not in [92.0, 95.0, 153.0, 158.0]
# condition 1 or condition 2

# condition 1 | condition 2
# (df['PT08.S2(NMHC)'] < 740) | (~df['NOx(GT)'].isin([92.0, 95.0, 153.0, 158.0]))
con_1 = df['PT08.S2(NMHC)'] < 740
con_2 = ~df['NOx(GT)'].isin([92.0, 95.0, 153.0, 158.0])
df.loc[con_1 | con_2][['Date','Time','T']]
```

#### Apply

The last common operation we saw was wanting to transform your data based on some function. For example, if we wanted to compute the length of each element in column **CO(GT)**, we could not run the following cell

```python
len(df['CO(GT)'])
```

This doesn't work because len returns the length of that whole column. You can use the apply function to do this! It's kind of strange at first, but it actually takes a function as a parameter! It will apply the given function to each element. For example, we can pass in the **len** function (notice we don't say **len()** because we don't want to actually call it).

```python
df['CO(GT)'].apply(len)
```

But this is actually more general since you can pass in any function that takes a single value and returns a new one!

```python
def first(s):
  return s[:2]

df['CO(GT)'].apply(first)
```

```python
# Exercise: multiply PT08.S2(NMHC) by 2.5
def mul(s):
  return s*2.5

df['PT08.S2(NMHC)'].apply(mul)
```

### Scatterplots

Scatterplots are usually used to explore the relationship between two continuous variables. Suppose now we wanna investigate the relationship between **NMHC(GT)** and **NO2(GT)**

```python
plt.rcParams["figure.figsize"] = (9,6) # set image size
plt.scatter(df['NMHC(GT)'], df['NO2(GT)'],  s = 40, color='black', alpha = 0.4) # scatterplot
plt.xlabel('NMHC(GT)') # x axis label
plt.ylabel('NO2(GT)') # y axis label
plt.show() # display the image
```

```python
# Exercise: plot the relationship between PT08.S4(NO2) and NO2(GT)
plt.rcParams["figure.figsize"] = (9,6) # set image size
plt.scatter(df['PT08.S4(NO2)'], df['NO2(GT)'],  s = 20, color='red', alpha = 0.9) # scatterplot
plt.xlabel('PT08.S4(NO2)') # x axis label
plt.ylabel('NO2(GT)') # y axis label
plt.show() # display the image
```
